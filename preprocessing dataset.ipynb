# Super Store Sales Analysis

```python
#import library
import pandas as pd
```

```python
#load the file
df=pd.read_csv("/Users/candicewu/Desktop/Superstore Dataset.csv")
df
```

```python
#check each column's type
df.info()
```

```python
# convert the 'Date' column to datetime format
df['Order Date']= pd.to_datetime(df['Order Date'])
df['Ship Date']= pd.to_datetime(df['Ship Date'])

# Check the format of 'Date' column
df.info()
```

```python
#sort values by Order Date column
df=df.sort_values(by="Order Date")
df
```

```python
# drop rows where ship date is earlier than order date
df = df[df['Ship Date'] >= df['Order Date']]
df
```

```python
#find the column with the number of missing values
df.isna().sum()
```

```python
#find the number of duplicates
df.duplicated().sum()
```

```python
# Show the duplicated rows
df[df.duplicated(keep = 'last')]
```

```python
# double check the duplicated rows
df[df.duplicated(keep = 'first')]
```

```python
#remove duplicates
df=df.drop_duplicates()
df.info()
```

```python
df.to_csv("/Users/candicewu/Desktop/cleaned superstore dataset.csv", index=False)
```
